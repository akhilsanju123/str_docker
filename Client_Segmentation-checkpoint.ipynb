```python
# Calculate the time difference between the first and last purchase for each customer
customer_dates = df.groupby('CustomerID')['Date']
customer_lifespans = (customer_dates.max() - customer_dates.min()).dt.days

# Adjust minimum lifespan to 1 day
customer_lifespans = customer_lifespans.where(customer_lifespans > 0, 1)

# Calculate the Average Customer Lifespan for each customer
average_customer_lifespans = customer_lifespans.groupby('CustomerID').mean()

# Calculate the Average Revenue per Customer for each customer
average_revenue_per_customer = df.groupby('CustomerID')['TotalSpend'].sum()

# Calculate CLV for Each Customer
clv_per_customer = average_revenue_per_customer * average_customer_lifespans
```

### RFM Segmentation

```python
# Define RFM scores
df_rfm = df.groupby('CustomerID').agg({
    'recency': 'min',
    'Frequency': 'sum',
    'TotalSpend': 'sum'
})

# Create labels for RFM scores
r_labels = range(4, 0, -1)
f_labels = range(1, 5)
m_labels = range(1, 5)

# Assign labels to RFM scores
df_rfm['R'] = pd.cut(df_rfm['recency'], bins=4, labels=r_labels)
df_rfm['F'] = pd.cut(df_rfm['Frequency'], bins=4, labels=f_labels)
df_rfm['M'] = pd.cut(df_rfm['TotalSpend'], bins=4, labels=m_labels)

# Combine RFM labels to create RFM segments
df_rfm['RFM_Segment'] = df_rfm['R'].astype(str) + df_rfm['F'].astype(str) + df_rfm['M'].astype(str)

# Calculate RFM score
df_rfm['RFM_Score'] = df_rfm[['R', 'F', 'M']].sum(axis=1)
```

```python
df_rfm.R = df_rfm.R.astype('int')
df_rfm.F = df_rfm.F.astype('int')
df_rfm.M = df_rfm.M.astype('int')
df_rfm.RFM_Segment = df_rfm.RFM_Segment.astype('int')
```

```python
# Calculate number of orders and total sales per day per customer
df_orders = df.groupby(['CustomerID', 'Date']).agg({'Quantity': 'sum', 'TotalSpend': 'sum'}).reset_index()

# Reset the index
df_orders = df_orders.reset_index(drop=True)

# Calculate number of orders and total sales over different time periods
df_orders['Orders_Last_Month'] = df_orders.groupby('CustomerID')['Quantity'].transform(lambda x: x.rolling(30, min_periods=1).sum()).reset_index(drop=True)
df_orders['Orders_Last_3_Months'] = df_orders.groupby('CustomerID')['Quantity'].transform(lambda x: x.rolling(90, min_periods=1).sum()).reset_index(drop=True)
df_orders['Orders_Last_6_Months'] = df_orders.groupby('CustomerID')['Quantity'].transform(lambda x: x.rolling(180, min_periods=1).sum()).reset_index(drop=True)
df_orders['Orders_Last_9_Months'] = df_orders.groupby('CustomerID')['Quantity'].transform(lambda x: x.rolling(270, min_periods=1).sum()).reset_index(drop=True)
df_orders['Orders_Last_12_Months'] = df_orders.groupby('CustomerID')['Quantity'].transform(lambda x: x.rolling(365, min_periods=1).sum()).reset_index(drop=True)

df_orders['Sales_Last_Month'] = df_orders.groupby('CustomerID')['TotalSpend'].transform(lambda x: x.rolling(30, min_periods=1).sum()).reset_index(drop=True)
df_orders['Sales_Last_3_Months'] = df_orders.groupby('CustomerID')['TotalSpend'].transform(lambda x: x.rolling(90, min_periods=1).sum()).reset_index(drop=True)
df_orders['Sales_Last_6_Months'] = df_orders.groupby('CustomerID')['TotalSpend'].transform(lambda x: x.rolling(180, min_periods=1).sum()).reset_index(drop=True)
df_orders['Sales_Last_9_Months'] = df_orders.groupby('CustomerID')['TotalSpend'].transform(lambda x: x.rolling(270, min_periods=1).sum()).reset_index(drop=True)
df_orders['Sales_Last_12_Months'] = df_orders.groupby('CustomerID')['TotalSpend'].transform(lambda x: x.rolling(365, min_periods=1).sum()).reset_index(drop=True)
```

```python
df_orders['Frequency_Last_Month'] = df_orders.groupby('CustomerID')['Date'].transform(lambda x: x.dt.floor('d').diff().lt(pd.Timedelta(days=30)).cumsum())
df_orders['Frequency_Last_3_Months'] = df_orders.groupby('CustomerID')['Date'].transform(lambda x: x.dt.floor('d').diff().lt(pd.Timedelta(days=90)).cumsum())
df_orders['Frequency_Last_6_Months'] = df_orders.groupby('CustomerID')['Date'].transform(lambda x: x.dt.floor('d').diff().lt(pd.Timedelta(days=180)).cumsum())
df_orders['Frequency_Last_9_Months'] = df_orders.groupby('CustomerID')['Date'].transform(lambda x: x.dt.floor('d').diff().lt(pd.Timedelta(days=270)).cumsum())
df_orders['Frequency_Last_12_Months'] = df_orders.groupby('CustomerID')['Date'].transform(lambda x: x.dt.floor('d').diff().lt(pd.Timedelta(days=365)).cumsum())
```
